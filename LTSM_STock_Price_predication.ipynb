```python
import pandas as pd
import datetime as dt
from datetime import date
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np
import tensorflow as tf
```

```python
START = "2010-01-01"
TODAY = date.today().strftime("%Y-%m-%d")
print(TODAY)

# Define a function to load the dataset

def load_data(ticker):
    data = yf.download(ticker, START, TODAY)
    data.reset_index(inplace=True)
    return data
```

```python
# Splitting data into training and testing

train = pd.DataFrame(data[0:int(len(data)*0.70)])
test = pd.DataFrame(data[int(len(data)*0.70): int(len(data))])

print(train.shape)
print(test.shape)
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
train_close = train.iloc[:, 4:5].values
test_close = test.iloc[:, 4:5].values
train_close
```

```python
data_training_array = scaler.fit_transform(train_close)
data_training_array.shape[0]
```

```python
x_train = []
y_train = [] 

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100: i])
    y_train.append(data_training_array[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train) 
```

```python
x_train[2].shape
# y_train[0]
```

```python
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(LSTM(units = 50, activation = 'relu', return_sequences=True
              ,input_shape = (x_train.shape[1], 1)))
model.add(Dropout(0.2))


model.add(LSTM(units = 60, activation = 'relu', return_sequences=True))
model.add(Dropout(0.3))


model.add(LSTM(units = 80, activation = 'relu', return_sequences=True))
model.add(Dropout(0.4))


model.add(LSTM(units = 120, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(units = 1))
```

```python
model.summary()
```

```python
import tensorflow as tf
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics=[tf.keras.metrics.MeanAbsoluteError()])
model.fit(x_train, y_train,epochs = 100)
```

```python
model.save('keras_model.h5')
print(test_close.shape)
past_100_days = pd.DataFrame(train_close[-100:])
test_df = pd.DataFrame(test_close)
final_df = past_100_days.append(test_df, ignore_index = True)
print(final_df.head())
input_data = scaler.fit_transform(final_df)
print(input_data)
print(input_data.shape)
```

```python
x_test = []
y_test = []
for i in range(100, input_data.shape[0]):
   x_test.append(input_data[i-100: i])
   y_test.append(input_data[i, 0])
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
print(x_test.shape)
print(y_test.shape)
```

```python
# Making predictions

y_pred = model.predict(x_test)
```

```python
scale_factor = 1/0.00041967
y_pred = y_pred * scale_factor
y_test = y_test * scale_factor
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
mae_percentage = (mae / np.mean(y_test)) * 100
print("Mean absolute error on test set: {:.2f}%".format(mae_percentage))
```

```python
from sklearn.metrics import r2_score

# Actual values
actual = y_test

# Predicted values
predicted = y_pred

# Calculate the R2 score
r2 = r2_score(actual, predicted)

print("R2 score:", r2)
```

```python
# Extract the Cell Gate weights
num_units = lstm_layer.units
cell_gate_start = 2 * num_units  # Start index for Cell Gate weights in W
cell_gate_end = 3 * num_units    # End index for Cell Gate weights in W

W_C = W[:, cell_gate_start:cell_gate_end]  # Weights for Cell Gate (third part)
U_C = U[2*num_units:3*num_units, :]  # Recurrent weights for Cell Gate (third part)
b_C = b[2*num_units:3*num_units]  # Biases for Cell Gate (third part)

# Print the Cell Gate weights
print("Cell Gate Weights (W_C):")
print(W_C)
print("Cell Gate Recurrent Weights (U_C):")
print(U_C)
print("Cell Gate Biases (b_C):")
print(b_C)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Create a Sequential model
model = Sequential()

# Add an LSTM layer
lstm_layer = LSTM(units=50, return_sequences=True, input_shape=(10, 1))
model.add(lstm_layer)

# Add a Dense output layer
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Build the model (this initializes the weights)
model.build(input_shape=(None, 10, 1))

# Access the weights of the LSTM layer
weights = lstm_layer.get_weights()

# Extract the weight matrix for the Forget, Input, Cell, and Output Gates
W = weights[0]  # Input weights
U = weights[1]  # Recurrent weights
b = weights[2]  # Biases

num_units = lstm_layer.units
input_dim = W.shape[0]  # Number of features in the input

# Index ranges for each gate
forget_gate_start = 0
forget_gate_end = num_units
input_gate_start = num_units
input_gate_end = 2 * num_units
cell_gate_start = 2 * num_units
cell_gate_end = 3 * num_units
output_gate_start = 3 * num_units
output_gate_end = 4 * num_units

# Extract weights for each gate
W_f = W[:, forget_gate_start:forget_gate_end]  # Weights for Forget Gate
W_i = W[:, input_gate_start:input_gate_end]  # Weights for Input Gate
W_C = W[:, cell_gate_start:cell_gate_end]  # Weights for Cell Gate
W_o = W[:, output_gate_start:output_gate_end]  # Weights for Output Gate

U_f = U[forget_gate_start:forget_gate_end, :]  # Recurrent weights for Forget Gate
U_i = U[input_gate_start:input_gate_end, :]  # Recurrent weights for Input Gate
U_C = U[cell_gate_start:cell_gate_end, :]  # Recurrent weights for Cell Gate
U_o = U[output_gate_start:output_gate_end, :]  # Recurrent weights for Output Gate

b_f = b[forget_gate_start:forget_gate_end]  # Biases for Forget Gate
b_i = b[input_gate_start:input_gate_end]  # Biases for Input Gate
b_C = b[cell_gate_start:cell_gate_end]  # Biases for Cell Gate
b_o = b[output_gate_start:output_gate_end]  # Biases for Output Gate

# Print the weights for each gate
print("Forget Gate Weights (W_f):")
print(W_f)
print("Forget Gate Recurrent Weights (U_f):")
print(U_f)
print("Forget Gate Biases (b_f):")
print(b_f)

print("\nInput Gate Weights (W_i):")
print(W_i)
print("Input Gate Recurrent Weights (U_i):")
print(U_i)
print("Input Gate Biases (b_i):")
print(b_i)

print("\nCell Gate Weights (W_C):")
print(W_C)
print("Cell Gate Recurrent Weights (U_C):")
print(U_C)
print("Cell Gate Biases (b_C):")
print(b_C)

print("\nOutput Gate Weights (W_o):")
print(W_o)
print("Output Gate Recurrent Weights (U_o):")
print(U_o)
print("Output Gate Biases (b_o):")
print(b_o)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Function to add a rectangle with label
def add_rectangle(ax, position, size, label, color='lightblue'):
    rect = patches.FancyBboxPatch(position, size[0], size[1], boxstyle="round,pad=0.3", edgecolor='black', facecolor=color)
    ax.add_patch(rect)
    plt.text(position[0] + size[0] / 2, position[1] + size[1] / 2, label, 
             horizontalalignment='center', verticalalignment='center', fontsize=12, weight='bold')

# Add rectangles for different components
add_rectangle(ax, (0.1, 0.8), (0.15, 0.1), 'Input 1', color='lightgreen')
add_rectangle(ax, (0.1, 0.65), (0.15, 0.1), 'Input 2', color='lightgreen')
add_rectangle(ax, (0.1, 0.5), (0.15, 0.1), 'Input 3', color='lightgreen')
add_rectangle(ax, (0.3, 0.65), (0.2, 0.1), 'Forget Gate\n(f)', color='lightcoral')
add_rectangle(ax, (0.3, 0.5), (0.2, 0.1), 'Input Gate\n(i)', color='lightcoral')
add_rectangle(ax, (0.55, 0.57), (0.2, 0.1), 'Cell State\n(C)', color='lightgoldenrodyellow')
add_rectangle(ax, (0.8, 0.57), (0.15, 0.1), 'Output Gate\n(o)', color='lightblue')
add_rectangle(ax, (0.3, 0.2), (0.35, 0.1), 'Output\n(h_t)', color='lightcyan')

# Add arrows for connections
def add_arrow(ax, start, end):
    arrow = patches.FancyArrowPatch(start, end, mutation_scale=15, arrowstyle='->', color='black')
    ax.add_patch(arrow)

# Draw arrows
add_arrow(ax, (0.25, 0.85), (0.3, 0.75))  # Input 1 -> Forget Gate
add_arrow(ax, (0.25, 0.7), (0.3, 0.6))  # Input 2 -> Forget Gate
add_arrow(ax, (0.25, 0.55), (0.3, 0.45))  # Input 3 -> Forget Gate
add_arrow(ax, (0.3, 0.7), (0.55, 0.6))  # Forget Gate -> Cell State
add_arrow(ax, (0.3, 0.55), (0.55, 0.55))  # Input Gate -> Cell State
add_arrow(ax, (0.55, 0.6), (0.8, 0.65))  # Cell State -> Output Gate
add_arrow(ax, (0.8, 0.65), (0.8, 0.75))  # Output Gate -> Output

# Add arrows for connections between gates and cell state
add_arrow(ax, (0.55, 0.57), (0.55, 0.55))  # Cell State input to Output Gate
add_arrow(ax, (0.3, 0.4), (0.55, 0.45))  # Input Gate -> Cell State update
add_arrow(ax, (0.3, 0.4), (0.55, 0.45))  # Cell State influence from Input Gate

# Add annotations
plt.text(0.1, 0.9, 'Inputs', fontsize=12, weight='bold')
plt.text(0.3, 0.75, 'Gates', fontsize=12, weight='bold')
plt.text(0.55, 0.75, 'Cell State', fontsize=12, weight='bold')
plt.text(0.8, 0.75, 'Output', fontsize=12, weight='bold')

# Set limits and remove axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Display the diagram
plt.title('LSTM Cell Diagram with Three Inputs')
plt.show()
```
